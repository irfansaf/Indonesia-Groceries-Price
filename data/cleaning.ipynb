```python
# Cleaning Pricing Data
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Load the data xlsx
data = pd.read_excel('harga.xlsx')

# Check the data
data.head()
```

```python
# Check the data type
data.dtypes
```

```python
# Check missing values and drop it
data.isnull().sum()

# Drop missing values
data = data.dropna()

# Check the data
data.isnull().sum()

# Check the data
data.head()
```

```python
date_columns = data.columns[2:]

for col in date_columns:
    try:
        # Format the date as you need
        new_col = pd.to_datetime(col, format='%d/ %m/ %Y').strftime('%Y-%m-%d')
        data.rename(columns={col: new_col}, inplace=True)
    except ValueError:
        print(f"Column {col} can't be converted to a date.")

# Print the new column names
print(data.columns)
```

```python
# Check all columns where there are "-" values
for col in data.columns:
    if data[col].dtype == 'O':
        print(f"Column {col} has {len(data[data[col] == '-'])} missing values.")
        
# Replace "-" with np.nan
data = data.replace('-', np.nan)

# Check the data
data.head()
```

```python
# Delete column if the entire column is nan
data = data.dropna(axis=1, how='all')

# Check the data
data.head()
```

```python
# Convert price columns to float start from column 2 and row 2
price_columns = data.columns[2:]

for col in price_columns:
    data[col] = data[col].str.replace(',', '').astype(float)
    
# Check the data
data.head()
```

```python
data.head()
```

```python
# Save the data
data.to_csv('harga_cleaned.csv', index=False)
data.to_excel('harga_cleaned.xlsx', index=False)
```
